```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(font='Malgun Gothic')
```

# 데이터 로드하기

```python
df_raw = pd.read_csv("C:/Users/82109/data_setting/data/국가_대륙_별_상품군별_온라인쇼핑_해외직접판매액_20230109162619.csv",encoding = "cp949")
df_raw.head()
```

```python
df_raw[df_raw['국가(대륙)별']== '미국']
```

# 분석과 시각화를 위한 tidy data

```python
df_raw.columns
```

```python
# 나머지 컬럼들이 variable값으로 내려왔어 (df.melt 함수 이용)
df = df_raw.melt(id_vars=['국가(대륙)별','상품군별','판매유형별'], var_name = '기간',value_name = '백만원')
df.head()
```

# 데이터 전처리

## 기간에서 연도를 분리하기

```python
df['연도'] = df['기간'].map(lambda x: int(x.split('.')[0]))
df['연도']
```

```python
df['분기'] = df['기간'].map(lambda x: int(x.split('.')[1].split('/')[0]))
df['분기']
```

```python
# 결측치들이 nan값 대신 '-'로 표현되어 있어
df['백만원'] = df['백만원'].replace('-',np.nan).astype(float)
df
```

```python
df.info()
```

```python
# 합계가 아닌 데이터로 다시 집계해
df = df[(df['국가(대륙)별'] != '합계')&(df['상품군별'] != '합계')].copy()
df
```

```python
# 용량이 줄어든 것을 확인할 수 있어
df.info()
```

```python
# Nan값 처리
df = df.dropna()
```

# K-Beauty 시각화

```python
# 합계데이터만 가져오기
df_total = df[df['판매유형별'] == '계'].copy()
df_total.head()
```

```python
sns.lineplot(data = df_total, x= '연도', y='백만원')
```

## relplot = subplot으로 시각화할 수 있게 해줘

```python
# 상품군 별로 다른 색상으로 표현하고 싶어 hue 이용
# legend 값을 밖에 표시하는 소스코드를 stack overflow에서 가져올 수 있어
sns.lineplot(data = df_total, x= '연도', y='백만원', hue = '상품군별')
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad = 0.)
```

```python
# 위에 그린 그래프를 서브플롯으로 표시하기
sns.relplot(data = df_total, x= '연도', y='백만원', hue = '상품군별', kind= 'line'
           , col ='상품군별', col_wrap =4)
```

```python
# 화장품제외 데이터 저장
df_sub = df_total[~df_total['상품군별'].isin(['화장품','의류 및 패션 관련상품'])].copy()
df_sub.head()
```

```python
sns.relplot(data = df_sub, x= '연도', y='백만원', hue = '상품군별', kind= 'line'
           , col ='상품군별', col_wrap =4)
```

## 화장품의 온라인쇼핑 해외직접판매액

```python
df_cosmetic = df_total[df_total['상품군별']=='화장품'].copy()
df_cosmetic['상품군별'].unique()
```

```python
# 연도와 판매액을 분기별로 다른 색상으로 표현해보기
sns.lineplot(data = df_cosmetic, x= '연도', y='백만원', hue = '분기')
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad = 0.)
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터 시각화하기
plt.figure(figsize = (15,4))
plt.xticks(rotation=30)
sns.lineplot(data = df_cosmetic, x= '기간', y='백만원', hue = '분기')
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad = 0.) 
```

```python
# 국가별로 데이터 시각화하기
plt.figure(figsize = (15,4))
plt.xticks(rotation=30)
sns.lineplot(data = df_cosmetic, x= '기간', y='백만원', hue = '국가(대륙)별')
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad = 0.) 
```

```python
# 중국만 빼고 한번 봐보자
plt.figure(figsize = (15,4))
plt.xticks(rotation=30)
sns.lineplot(data = df_cosmetic[df_cosmetic['국가(대륙)별'] != '중국'], x= '기간', y='백만원', hue = '국가(대륙)별')
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad = 0.) 
```

```python
# 화장품 판매액에 대한 기간별 금액 데이터시각화하고 '판매유형별'로 다른 색상으로 표시하기
df_sub = df[df['판매유형별'] != '계'].copy()
plt.figure(figsize = (15,4))
plt.xticks(rotation=30)
sns.lineplot(data = df_sub, x= '기간', y='백만원', hue = '판매유형별')
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad = 0.) 
```

## 의류 및 패션 관련상품 온라인쇼핑 해외직접판매액

```python
# 패션 데이터가져오기
df_fashion = df[(df['상품군별'] == '의류 및 패션 관련상품')&(df['판매유형별']=='계')].copy()
df_fashion.head()
```

```python
#국가(대륙)별 다른 색상으로 보기
plt.figure(figsize = (15,4))
plt.xticks(rotation=30)
sns.lineplot(data = df_fashion, x= '기간', y='백만원', hue = '국가(대륙)별')
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad = 0.) 
```

```python
# 계 제거하고 그리기
df_fashion_2 = df[(df['상품군별'] == '의류 및 패션 관련상품')&(df['판매유형별']!='계')].copy()
df_fashion_2.head()
```

```python
# 코로나 이후 면세점 판매량이 확 감소한 것을 확인할 수 있어
plt.figure(figsize = (15,4))
plt.xticks(rotation=30)
sns.lineplot(data = df_fashion_2, x= '기간', y='백만원', hue = '판매유형별',ci= None)
plt.legend(bbox_to_anchor=(1.05,1),loc=2,borderaxespad = 0.) 
```
